```python
import pandas as pd
from pycaret.classification import *
```

```python
df = pd.read_csv('/content/Train_dataset.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
cat_features=['sex','cp','fbs','restecg','exng','slp','caa','thall']
```

```python
experiment=setup(data=df, target='output',
      remove_outliers = True, outliers_threshold = 0.05,
      normalize = True, normalize_method = 'zscore',
      transformation = True, transformation_method = 'yeo-johnson',categorical_features=cat_features)
```

```python
best=compare_models()
```

```python
predict_model(best)
```

```python
predict_model(best,df.tail())
```

```python
save_model(best,model_name='decision_tree_classifier')
```

```python
model=load_model('decision_tree_classifier')
```

```python
new_df=pd.read_csv('/content/test_dataset.csv')
new_df
```

```python
prediction= predict_model(model,data=new_df)
```

```python
print(prediction)
```

```python
prediction.rename(columns={'prediction_label': 'output'}, inplace=True)
```
